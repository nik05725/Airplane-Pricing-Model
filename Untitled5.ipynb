```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Clean_Dataset.csv')
```

```python
df.sample(10)
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df['class'].value_counts()
```

```python
import seaborn as sns
sns.scatterplot(data=df, x='price', y='airline',hue='class')
```

```python
df.duration.median()
```

```python
df['class']=df['class'].map({'Economy':1,'Business':0})
```

```python
df
```

```python
df['stops']=df['stops'].map({'zero':0,'one':1})
```

```python
df=df.drop('flight', axis=1)
```

```python
df =df.join(pd.get_dummies(df.source_city, prefix='source_')).drop('source_city', axis=1)
df =df.join(pd.get_dummies(df.destination_city, prefix='destination_')).drop('destination_city', axis=1)
df =df.join(pd.get_dummies(df.arrival_time, prefix='arrival_')).drop('arrival_time', axis=1)
df =df.join(pd.get_dummies(df.departure_time, prefix='departure_')).drop('departure_time', axis=1)
df =df.join(pd.get_dummies(df.airline, prefix='airline_')).drop('airline', axis=1)
```

```python
df
```

```python
df=df.astype(int)
```

```python
df = df.dropna()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X,y=df.drop('price', axis=1), df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
reg = RandomForestRegressor(n_jobs=-1)
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
import math
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
y_pred = reg.predict(X_test)

print('R2:', r2_score(y_test,y_pred))
print('MAE:', mean_absolute_error(y_test,y_pred))
print('MSE:', mean_squared_error(y_test,y_pred))
print('RMSE:', math.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Flight Price')
plt.ylabel('Predicted Flight Price')
```
